# filter data with the Query() method

```python
# select players older than 34 from italy  &== and
pf.query("age>34 and nationality=='Italy'")
# exercise: write the equivalent boolean slicing
pf(pf['age']>34 )& ( nationality=='Italy')]
```

```python
#add a not operator to the first exanple
pf.query("not(age>34)")
#excercise: write the equivalent boolean slicing
pf[~(pf['age']>34)]
```

```python
#convert height to meters and select thise with height above 1.8
pf.query("height_cm /100 > 1.8")
# exercise : write the equivalent boolean slicing
pf[(pf['height_cm'] / 100) > 1.8]
```

## 1 copy()

#### 1.1 deep = True

```python
# deep = True by default  (modifications to the data or indices pf the copy)
pf_copy =pf.copy()
```

```python
# update value in original dataframe
pf.loc['L. Messi','height_cm'] = 190
```

```python
# copy vs original dataframe
pf
pf_copy
```

## 1.2 deep = False

```python
# deep =false ( any changes to the data of the original will be reflected in the shallow copy and vise versa)
pf_copy =pf.copy(deep = False)
```

```python
# update value in original dataframe
pf.loc['Cristiano Ronaldo','height_cm'] = 178
```

```python
# shallow copy vs original dataframe
pf
pf_copy
```
