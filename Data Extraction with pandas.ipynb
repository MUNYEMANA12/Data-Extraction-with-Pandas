```python
import pandas as pd
import numpy as np
```

```python
# read csv file
pf = pd.read_csv('players_20.csv')
```

```python
# show dataframe
pf
```

```python
# set index
pf.set_index('short_name',inplace= True)
```

```python
# show dataframe
pf
```

```python
# select columns
pf= pf[['long_name','age','dob','height_cm','weight_kg','nationality','club']]
```

```python
pf
```

# 1. selecting with a single value

#### loc[row,column_label]

```python
# get all data about  L.Messi
pf.loc['L. Messi']
```

```python
# get the height of L.messi
pf.loc['L. Messi','height_cm']
```

```python
# get the weight of cristiano  Ronaldo
pf.loc['Cristiano Ronaldo','weight_kg']
```

```python
# get all rows inside the "height_cm" column
pf.loc[:,'height_cm']
```

```python
# to get all columns that corresponded to the index 'L. Messi'
pf.loc['L. Messi',:]
```

# 2. Selecting with a list of values  with loc[] Method

```python
# get all data about L.Messi and Cristiano Ronaldo
pf.loc[['L. Messi','Cristiano Ronaldo']]
```

```python
# get height of L.Messi nad Cristiano Ronaldo
pf.loc[['L. Messi','Cristiano Ronaldo'],'height_cm']
```

```python
# get the height and weight of L.Messi
pf.loc['L. Messi',['height_cm','weight_kg']]
```

```python
# get the height and weight of L.Messi and Cristiano Ronaldo
pf.loc[['L. Messi','Cristiano Ronaldo'],['height_cm','weight_kg']]
```

# 3. Selecting  a range of data with a slice

#### start: stop: step (note that contrary to usual python slices,both the start and the stop are included)

```python
# slice column labels
players= ['L. Messi','Cristiano Ronaldo']

pf.loc[players,'age': 'club']
```

```python
# slice index labels
columns=['age','dob','height_cm','weight_kg','nationality','club']

#get top1 and top10 player name
pf.index[:10]
pf.loc['L. Messi' : 'M. Salah',columns]
```

# 4. selsecting with conditions

```python
# one conditions: select player with height above 180cm
columns =['age','dob','height_cm','weight_kg']
pf.loc[pf['height_cm'] > 180,columns]
```

```python
# multiple conditions : select players with height above 180cm frm Argentina
pf['height_cm'] > 180 
pf['nationality'] == 'Argentina'

pf.loc[(pf['height_cm'] > 180) & (pf['nationality'] == 'Argentina'),:]
```
